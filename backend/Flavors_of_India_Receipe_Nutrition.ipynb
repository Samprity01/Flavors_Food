```python
import pandas as pd
import matplotlib.pyplot as plt

class NutritionalAnalysis:
    def __init__(self):
        self.nutritional_data = {}

    def load_nutritional_data(self, ingredient_data):
        """
        Load nutritional data dynamically.

        :param ingredient_data: A dictionary containing ingredient names and their nutritional values.
        """
        self.nutritional_data = ingredient_data

    def analyze_recipe(self, recipe_name, ingredients):
        """
        Analyze the nutritional content of the given recipe ingredients.

        :param recipe_name: The name of the recipe.
        :param ingredients: A list of ingredient names.
        :return: A dictionary containing total nutritional values.
        """
        total_nutrition = {'recipe_name': recipe_name, 'calories': 0, 'protein (g)': 0, 'fat (g)': 0, 'carbs (g)': 0}

        for ingredient in ingredients:
            if ingredient in self.nutritional_data:
                # Assume the quantity is for simplicity; you can modify this to accept dynamic quantities
                quantity = 100  # Default quantity in grams
                nutrition = self.nutritional_data[ingredient]

                total_nutrition['calories'] += nutrition['calories']
                total_nutrition['protein (g)'] += nutrition['protein'] * (quantity / 100)
                total_nutrition['fat (g)'] += nutrition['fat'] * (quantity / 100)
                total_nutrition['carbs (g)'] += nutrition['carbs'] * (quantity / 100)

        return total_nutrition

    def plot_nutrition(self, nutrition_info):
        """
        Plot the nutritional information in a bar graph.

        :param nutrition_info: A dictionary containing nutritional values.
        """
        categories = ['Calories', 'Protein (g)', 'Fat (g)', 'Carbs (g)']
        values = [nutrition_info['calories'], nutrition_info['protein (g)'], nutrition_info['fat (g)'], nutrition_info['carbs (g)']]

        plt.figure(figsize=(8, 5))
        plt.bar(categories, values, color=['orange', 'green', 'blue', 'red'])
        plt.xlabel('Nutritional Components')
        plt.ylabel('Amount per 100g')
        plt.title(f'Nutritional Breakdown of {nutrition_info["recipe_name"]}')
        plt.grid(axis='y')
        plt.show()

# Example usage
if __name__ == "__main__":
    # Dynamic ingredient data input
    dynamic_nutritional_data = {
        'rice': {'calories': 130, 'protein': 2.7, 'fat': 0.3, 'carbs': 28.2},
        'chicken': {'calories': 239, 'protein': 27.3, 'fat': 13.6, 'carbs': 0},
        'spices': {'calories': 100, 'protein': 3.5, 'fat': 5.0, 'carbs': 20.0},
        'yogurt': {'calories': 59, 'protein': 10, 'fat': 0.4, 'carbs': 3.6},
        'cucumber': {'calories': 16, 'protein': 0.7, 'fat': 0.1, 'carbs': 3.6},
        'paneer': {'calories': 265, 'protein': 18.6, 'fat': 20.0, 'carbs': 1.2},
        'bell pepper': {'calories': 20, 'protein': 0.9, 'fat': 0.2, 'carbs': 4.7},
        'flour': {'calories': 364, 'protein': 10.3, 'fat': 1.0, 'carbs': 76.3},
        'potatoes': {'calories': 77, 'protein': 2.0, 'fat': 0.1, 'carbs': 17.6},
        'milk': {'calories': 42, 'protein': 3.4, 'fat': 1.0, 'carbs': 5.0},
        'sugar': {'calories': 387, 'protein': 0, 'fat': 0, 'carbs': 100}
    }

    # Initialize the nutritional analysis system
    analysis_system = NutritionalAnalysis()

    # Load dynamic nutritional data into the system
    analysis_system.load_nutritional_data(dynamic_nutritional_data)

    # Get user input for recipe name and ingredients
    recipe_name = input("Enter the recipe name: ")
    ingredients_input = input("Enter the ingredients separated by commas: ")

    # Process user input to create a list of ingredients
    recipe_ingredients = [ingredient.strip() for ingredient in ingredients_input.split(',')]

    # Analyze the recipe
    nutritional_info = analysis_system.analyze_recipe(recipe_name, recipe_ingredients)

    print("Nutritional Breakdown:")
    print(nutritional_info)

    # Plot the nutritional information
    analysis_system.plot_nutrition(nutritional_info)
```
